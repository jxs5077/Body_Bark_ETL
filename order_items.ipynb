```python
test_str=df_filtered.iloc[0, 7]
display(test_str)
def extract_product(product_detail_str): 
    return_list=[]
    product_list=product_detail_str.split('|')
    for each_product in product_list:
        product_dict={}
        product_detail=each_product.split(', Product')
        product_dict['total_each_sku']=product_detail[1].split(': ')[1]
        product_dict['sku']=product_detail[2].split(': ')[1]
        product_dict['product_unit_price']=product_detail[6].split(': ')[1]
        return_list.append(product_dict)
    return return_list
```

```python
extract_product(test_str)
```

```python
order_product_details=[]
for idx, each_row in df_filtered.iterrows(): 
    order_product_list=extract_product(each_row['Product Details'])
    for each_product in order_product_list: 
        each_product['order_id']=each_row['order_id']
        order_product_details.append(each_product)
new_df=pd.DataFrame(order_product_details)
```
