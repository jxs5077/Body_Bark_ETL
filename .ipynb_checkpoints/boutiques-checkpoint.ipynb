```python
import pandas as pd
```

```python
df_name = pd.read_csv('Resources/Body_Bark_Customer_list_bill_address.csv')
```

```python
type(df_name["zip"])
```

```python
df_name[['zip_code','extra']] = df_name.zip.str.split(pat='-', expand=True)
```

```python
#for i in range(len([df_name.zip])):
df_name['zip_code']=df_name['zip_code'].apply(lambda x: '{0:0>5}'.format(x))  
```

```python
type(df_name['zip_code'][0])
```

```python
df_filtered = df_name.drop(['zip', 'extra'], axis=1)  #axis changes from row to column
```

```python
boutique_customers_df = df_filtered.rename(columns={"zip_code": "zip"})
```

```python
boutique_customers_df.reset_index(inplace=True)
```

```python
boutique_customers_df = boutique_customers_df.rename(columns={"index": "boutique_id"})
```

```python
column_names = ['boutique_id', 'boutique_name', 'street_address', 'city', 'state', 'zip']
boutique_customers__columnorder_df = boutique_customers_df.reindex(columns=column_names)
```

```python
boutique_customers__columnorder_df.head(30)
```

```python
boutique_customers__columnorder_df.to_csv('output/Boutiques.csv', index = False)
```
