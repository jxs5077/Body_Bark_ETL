```python
import pandas as pd
```

```python
df_name = pd.read_csv('Resources/Body_Bark_ProductService_list.csv')
```

```python
df_name.head(20)
```

```python
df_name[['product_num','product_color', 'product_size']] = df_name.sku.str.split(pat='-', expand=True)
```

```python
df_name.head(20)
```

```python
product_df=df_name
```

```python
product_df.to_csv('output/Product_List.csv', index = False)
```

```python
from sqlalchemy import create_engine
conn='postgres://postgres:password@localhost:5432/BodyBark'
engine=create_engine(conn)
```

```python
engine.table_names()
```

```python
product_df.to_sql('Product_List', if_exists='append', con=engine, index=False)
```
